```python
from sklearn.cluster import k_means
from sklearn import datasets
nosify_moons,_ = datasets.make_moons(n_samples=200, shuffle=True, noise=0.05, random_state=33)
nosify_moons
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(nosify_moons[:,0],nosify_moons[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans_c = KMeans(n_clusters=2).fit_predict(nosify_moons)
kmeans_c
```

```python
plt.scatter(nosify_moons[:,0],nosify_moons[:,1],c=kmeans_c,cmap='bwr') #通过kmeans发现并不是我们想要的聚类
plt.show()
```

```python
from sklearn.cluster import DBSCAN #导入dbscan算法
dbscan_sk = DBSCAN(eps=0.3, min_samples=5, metric='euclidean') #eps的大小将影响到找到簇的数量
dbscan_sk_c = dbscan_sk.fit_predict(nosify_moons) #拟合预测聚簇中心
print(nosify_moons)
print(dbscan_sk_c)
```

```python
plt.scatter(nosify_moons[:,0],nosify_moons[:,1],c=dbscan_sk_c,cmap='bwr') #通过kmeans发现并不是我们想要的聚类
plt.show()
```
