```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt # 可视化
import numpy as np
blobs,_ = make_blobs(n_samples=300,n_features=2,centers=3,random_state=10) #samples样本数，features 特征数，centers中心数(聚簇中心)
blobs.shape
```

```python
#构造一个函数
def random_k(k,data):   #初始化质心点 k=数量,多少个质心数 data=传入的数据集
    #初始化
    prng = np.random.RandomState(27)
    num_features = np.shape(data)[1] #获取特征数量
    init_centers = prng.randn(k,num_features)
    return init_centers
init_centers = random_k(3,blobs)
```

```python
plt.scatter(blobs[:,0],blobs[:,1],s=20)
plt.scatter(init_centers[:,0]*5,init_centers[:,1]*17,s=30,marker="*",c='r')
plt.show()
```

```python
from sklearn.cluster import k_means #导入kmeans算法
```

```python
model = k_means(blobs,n_clusters=3)
print("初始质心",init_centers)
print("聚类之后得到的质心",model[0])
```

```python
plt.scatter(blobs[:,0],blobs[:,1],s=20)
plt.scatter(model[0][:,0],model[0][:,1],s=30,marker="*",c='r')
plt.show()
```
